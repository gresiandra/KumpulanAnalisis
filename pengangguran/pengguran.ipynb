# Analisis Data Pengangguran di Indonesia

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('Pengangguran_1986-2018.csv', sep=';')
data
```

```python
data.shape
```

# Transform Data (,) to (.)

```python
def toFloat(angka):
    i = 0
    for row in data[' '+str(angka)+' ']:
        row = row.replace(' ', '')
        row = row.replace(',', '.')    
        data.loc[i, ' '+str(angka)+' '] = row
        i+=1
```

# Transform Data to Float & Empty Data to NaN

```python
def toNaN(angka):
    i= 0
    for row in data[' '+str(angka)+' ']:  
        try:
            row = float(row)
        except ValueError:
            data.loc[i, ' '+str(angka)+' '] = np.nan 
        i+=1
```

# Using Transform Function

```python
angka = 1986

while(angka<2019):
    toFloat(angka)
    toNaN(angka)
    angka+=1
```

# Transform Data Result

```python
data
```

# Transform Column Header to Float 

```python
kolom = []

for column in data.columns:
    if column != 'Provinsi':
        kolom.append(column)
        
del kolom[0]
print(kolom)

data[kolom] = data[kolom].apply(pd.to_numeric)
```

```python
data.info()
```

# Filling NaN Data With Mean

```python
for col in kolom:
    mean = np.mean(data[col])
    data[col].fillna(mean, inplace=True)
```

```python
del data[' 1995 ']
```

```python
np.sum(data.isnull())
```

# Visualization Analytics

## Jakarta vs Kalimantan Timur

```python
jakarta = data[data[' Provinsi '] == ' DKI Jakarta ']
jakarta = jakarta.T

jakarta = jakarta[1:] #take the data less the header row
jakarta.columns = ['persentase']
jakarta
```

```python
kaltim = data[data[' Provinsi '] == ' Kalimantan Timur ']
kaltim = kaltim.T

kaltim = kaltim[1:]
kaltim.columns = ['persentase']
kaltim
```

```python
plt.plot(jakarta.index, jakarta.persentase)
plt.plot(kaltim.index, kaltim.persentase)
plt.xlabel('Tahun',fontsize=12)
plt.ylabel('Persentase Pengangguran',fontsize=12)
plt.xticks(rotation=90, fontsize=10)
plt.legend(['Jakarta','KalTim'])
plt.show()
```
