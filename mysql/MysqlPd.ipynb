# Pandas Dataframe to Mysql

```python
import pandas as pd
import pymysql
from sqlalchemy import create_engine
import matplotlib.pyplot as plt
import seaborn as sns
```

## Create mysql connection

```python
engine = create_engine('mysql+pymysql://root:@localhost/mtcars').connect()
```

## Make dataframe

```python
df = pd.read_csv('mtcars.csv')
df
```

## Insert dataframe to Mysql

```python
try:
    df.to_sql(con=engine, name='insert', if_exists='replace')
    print('Berhasil Memasukan Data')
except:
    print('Gagal Memasukan Data')
```

## Sedikit analisis menggunakan heatmap

```python
correlations = df.corr(method="pearson")

f,ax = plt.subplots(figsize=(12,12))
sns.heatmap(correlations, annot=True, linewidths=0.1, fmt='.3f', ax=ax)
plt.show()
```

```python
correlations = df.corr(method="spearman")

f,ax = plt.subplots(figsize=(12,12))
sns.heatmap(correlations, annot=True, linewidths=0.1, fmt='.3f', ax=ax)
plt.show()
```
